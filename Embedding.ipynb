```python
# This file consists of embedding layer which shows you how to use embedding technique and convert the text to numberic value. 
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# Sentences 
text = [ 'the glass of milk',
         'the glass of juice',
         'the cup of tea',
         'i am a good boy',
         'i am a good developer',
         'understands the meaning of words',
         'your vedios are good']
```

```python
# converting the words to vectors. 
voc_size = 10000
```

```python
# one hot representation
one_hot_rep = [one_hot(words, voc_size) for words in text]
one_hot_rep
# These numbers explain the index of each words mentioned in the text. 
```

```python
# Word embeddings representation 

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
# there will be 8 words in the sentence 
sent_length = 8
Embedding_docs= pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
print(Embedding_docs)
```

```python
# Feature representation 
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(Embedding_docs)
```
